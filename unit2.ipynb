```python
#pre-proccesing techinque
#Mean removal 
import numpy as np 

data = np.array([[150],[160],[170]])

mean = np.mean(data)

mean_removal  = data - mean 

print("orignal data",data)
print("mean data",mean)
print("mean removal",mean_removal)
```

```python
#scalling 

data = [50,100,150]

min_val = min(data)
max_val = max(data)

scaled_data = [(x - min_val)/(max_val - min_val) for x in data]

print(scaled_data)
```

```python
#normaliztion 
#normalization convert data into samll range  in  0 nad 1 

data = [50,100,150]

min_val = min(data)
max_val = max(data)

normalized_data = [(x - min_val)/(max_val - min_val) for x in data]

print(normalized_data)
```

```python
#Regression 
#linear regression 

from sklearn.linear_model import LinearRegression
import numpy as np 

x = np.array([[1],[2],[3],[4]]) #hours of study
y = np.array([20,40,60,80]) #marks

model = LinearRegression()

model.fit(x,y)

predict = model.predict([[5]])
print(predict)
```

```python
#ddecision tree regression 

from sklearn.tree import DecisionTreeRegressor
import numpy as np 

x = np.array([[1],[2],[3],[5]]) #years of experinece
y = np.array([40000,450000,50000,70000]) #salary

model = DecisionTreeRegressor()

model.fit(x,y)

predict = model.predict([[4]])
print(predict)
```

```python
#svm for regression 

from sklearn.svm import SVR
import numpy as np 

x = np.array([[1],[2],[3],[4],[5]]) #area
y = np.array([10,20,30,40,50])  #price

model = SVR(kernel='linear')

model.fit(x,y)

predict = model.predict([[6]])
print(predict)
```

```python
#classification 
#logistic regression

from sklearn.linear_model import LogisticRegression
import numpy as np 

x = np.array([[1],[2],[3],[4],[5]]) #stdeunt
y = np.array([0,0,1,1,1]) #0 for fail 1 for pass

model = LogisticRegression()

model.fit(x,y)

predict = model.predict([[4]])
print(predict)
```

```python
#navie byes alogoritham 

from sklearn.naive_bayes import MultinomialNB
import numpy as np 

x = np.array([[0,0],[0,1],[1,1],[1,0]]) #0 menase freee #1 means has offfer
y = np.array(['spam','spam','not spam','spam'])

model = MultinomialNB()

model.fit(x,y)
predict = model.predict([[1,1]])
print(predict)
```
