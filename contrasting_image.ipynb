```python
#contrastign the color image
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('rose.jpg')

#convert image to Ycrcb
ycrcb=cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)

#split the color
y,cr,cb = cv2.split(ycrcb)

#global histogram equalization
y_eq = cv2.equalizeHist(y)
ycrcb_eq = cv2.merge([y_eq,cr,cb])

#convert an image to bgr
global_eq = cv2.cvtColor(ycrcb_eq,cv2.COLOR_YCrCb2BGR)

#convert an image from bgr to rgb
source = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
global_eq1 = cv2.cvtColor(global_eq,cv2.COLOR_BGR2RGB)

#ploating the image
plt.subplot(1,2,1),plt.imshow(img),plt.title('Source image'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(global_eq1),plt.title('EqualizedHiwst on color image'),plt.axis('off')
```
