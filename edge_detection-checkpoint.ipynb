```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('tanjiro.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

size=30
k1=np.zeros((size,size))
k1[int(size/2),:]=np.ones(size)
k1=k1/size

horizontal_image = cv2.filter2D(img,-1,k1)
plt.figure(figsize=(10,6))
plt.subplot(1,2,1),plt.imshow(img),plt.title('Orignal image')
plt.subplot(1,2,2),plt.imshow(horizontal_image),plt.title('horizontal blur')
```

```python
#sobel

import cv2
import matplotlib.pyplot as plt
import numpy as np

img =cv2.imread('tiger.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,10)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,10)

soble_combined = cv2.magnitude(sobelx,sobely)

plt.figure(figsize=(10,6))
plt.subplot(1,4,1),plt.imshow(img),plt.title('Orignal image'),plt.axis('off')
plt.subplot(1,4,2),plt.imshow(sobelx,cmap='gray'),plt.title('sobelx image'),plt.axis('off')
plt.subplot(1,4,3),plt.imshow(sobely,cmap='gray'),plt.title('sobely image'),plt.axis('off')
plt.subplot(1,4,4),plt.imshow(soble_combined,cmap='gray'),plt.title('soble_combined image'),plt.axis('off')
```

```python
#laplacian
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('tiger.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#blur
gass = cv2.GaussianBlur(img,(3,3),0)

#laplacian
lap=cv2.Laplacian(gass,cv2.CV_64F,ksize=3)
cvt = cv2.convertScaleAbs(lap)
plt.figure(figsize=(10,6))
plt.subplot(1,2,1),plt.imshow(gass,cmap='gray'),plt.title('gaussuan blur'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(lap,cmap='gray'),plt.title('laplaican image'),plt.axis('off')
```

```python
#cany

import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('tiger.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

gas= cv2.GaussianBlur(img,(3,3),0)

#canny
canny = cv2.Canny(gas,threshold1=50,threshold2=150)

plt.figure(figsize=(10,6))
plt.subplot(1,2,1),plt.imshow(gas,cmap='gray'),plt.title('Gaussain blur'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(canny,cmap='gray'),plt.title('Caany edge detection'),plt.axis('off')
```

```python
#image sharpning

import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread('arial.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

k1=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])

sharp = cv2.filter2D(img,-1,k1)
plt.subplot(1,2,1),plt.imshow(img),plt.title('orignal image'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(sharp),plt.title('shrper image'),plt.axis('off')
```
