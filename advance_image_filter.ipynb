```python
#advance image filter
# (1) bilateral 
import cv2 
import matplotlib.pyplot as plt

img = cv2.imread('rose.jpg')
bila = cv2.bilateralFilter(img,d=9,sigmaColor=95,sigmaSpace=95)

plt.subplot(1,2,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)),plt.title('source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(cv2.cvtColor(bila,cv2.COLOR_BGR2RGB)),plt.title('smoothing the image'),plt.axis('off')
```

```python
#2 unsharp
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('aerial.jpg')

gauss = cv2.GaussianBlur(img,(9,9),10)

unsharp = cv2.addWeighted(img,1.5,gauss,-0.5,0)

plt.subplot(1,2,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)),plt.title('source'),plt.axis('off')
plt.subplot(1,2,2),plt.imshow(cv2.cvtColor(unsharp,cv2.COLOR_BGR2RGB)),plt.title('sharper the image'),plt.axis('off')
```

```python
#3 scharr (edge detection)
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('tiger1.jpg',0)

scharrx = cv2.Scharr(img,cv2.CV_64F,1,0)
scharry = cv2.Scharr(img,cv2.CV_64F,0,1)

scharrx = cv2.convertScaleAbs(scharrx)
scharry = cv2.convertScaleAbs(scharry)

#combine
scharr_combine = cv2.addWeighted(scharrx,0.10,scharry,0.2,0)

plt.subplot(1,3,1),plt.imshow(scharrx,cmap='gray'),plt.title('vertical edges'),plt.axis('off')
plt.subplot(1,3,2),plt.imshow(scharry,cmap='gray'),plt.title('vertical edges'),plt.axis('off')
plt.subplot(1,3,3),plt.imshow(scharr_combine,cmap='gray'),plt.title('vertical edges'),plt.axis('off')
```
